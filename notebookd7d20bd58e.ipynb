```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/electric-vehicle-population/Electric_Vehicle_Population_Data.csv')
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(10, 6))


sns.countplot(x='Electric Vehicle Type', data=data, order=data['Electric Vehicle Type'].value_counts().index)


plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability


plt.show()
```

```python

plt.figure(figsize=(10, 6))


sns.scatterplot(x='Electric Range', y='Base MSRP', data=data)


plt.title('Electric Range vs. Base MSRP')
plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')


plt.show()
```

```python
state_counts = data['State'].value_counts()
plt.figure(figsize=(10, 6))
state_counts.plot(kind='bar', color='skyblue')
plt.title('State-wise Electric Vehicle Adoption')
plt.xlabel('State')
plt.ylabel('Number of Electric Vehicles')
plt.xticks(rotation=45)
plt.show()
```
